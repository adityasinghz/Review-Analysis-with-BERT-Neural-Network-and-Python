```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification #AutoToken Converts String from sequence of numbers to pass in NLP model and Auto model For SequenceClassification ensures that transformers able to be load in NLP model
import torch   #extract high sequence results
import requests #grab data
from bs4 import BeautifulSoup #web scraping purposes to pull the data out of HTML files.
import re #regex which will be used to extact comments that we actually want                   
```

# Initialization Of Model

```python
tokenizer= AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment') #loading data in tokens
model= AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment') #loading tokens into the model
```

# Encode and Calculate Sentiment

```python
tokens = tokenizer.encode('I played this right when season 1 came out, and picked it up again after a few years. Gameplay seems to be way more varied than it used to be, and this game plays really well, especially as a mobile game. My two main gripes are: where did Zombies go? Did it go away for the season? Also, it boasts controller compatibility, but it has a hard time with connecting, even if the controller is working on the home screen of your phone, even on my husbands newer phone. All and all, pretty good game.', return_tensors='pt')#encoding the string
```

```python
result=model(tokens)
```

```python
result.logits
```

```python
int(torch.argmax(result.logits))+1    #review of upper comment
```

# Collect Reviews 

```python
r=requests.get('https://www.yelp.com/biz/r-and-g-lounge-san-francisco')  #collecting comments from this website
soup= BeautifulSoup(r.text,'html.parser')                                #scraping all the data from html file
regex=re.compile('.*comment.*')                                          #but we need only comment part thats why we've used regex
results=soup.find_all('p',{'class':regex})                               #the comment part is in class of the html code using regex I'm scraping comment part from class (which was extracted by beautifulsoup)
reviews=[result.text for result in results]                              #storing the reviews in the review variable
```

# Loading Reviews into DataFrames and Store

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.array(reviews), columns=['review'])
```

```python
df.head()
```

```python
def sentiment_score(review):
    tokens = tokenizer.encode(review, return_tensors='pt')
    result=model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
sentiment_score(df['review'].iloc[0])
```

```python
df['sentiment']=df['review'].apply(lambda x: sentiment_score(x[:512])) #at a time Neural netwoks can store only 512 data
```
